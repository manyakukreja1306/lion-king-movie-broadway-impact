Using regex to extract prices**

```python
price_split = df['Avg TicketTop Ticket'].str.findall(r'\$\d+(?:\.\d+)?')

df['Avg Ticket'] = price_split.str[0]
df['Top Ticket'] = price_split.str[1]

```

```python
seat_split = df['Seats SoldSeats in Theatre'].str.findall(r'\d[\d,]*')

df['Seats Sold'] = seat_split.str[0]
df['Seats in Theatre'] = seat_split.str[1]
```

```python
clean_seats = df['Seats SoldSeats in Theatre'].str.replace(',', '', regex=False)
```

```python
df['Seats in Theatre'] = pd.to_numeric(
    clean_seats.str[-4:], errors='coerce'
)
```

```python
df['Seats Sold'] = pd.to_numeric(
    clean_seats.str[:-4], errors='coerce'
)
```

**Conclusion and Justification**

(Week Ending)
Column begins with a header followed by valid weekly values. Since the data is correctly structured, this column can be used as-is after excluding the header row. It provides the temporal reference for all other metrics and is therefore retained without modification.

(Avg Ticket – Top Ticket)
Column is incorrectly labelled due to a repeated HTML class (col-1) in the source. Although the label is wrong, the underlying text contains two distinct values:
Average ticket price
Top ticket price

(Seats Sold – Seats in the Theatre)
Column is similarly mislabelled and contains two metrics combined into a single field:
Seats sold
Seats in the theatre
Initial inspection showed that the “Seats in the Theatre” value remained constant for a large portion of the dataset, which initially suggested that it might be safely omitted. However, further inspection across later weeks revealed that this value does change over time, meaning it represents meaningful information rather than redundant data.
Due to the absence of a reliable delimiter in the CSV (values were concatenated), the seats sold value was extracted directly, and the seats-in-theatre value was reconstructed only after validating its consistency and change across weeks. This ensures that no meaningful capacity information is lost.

(% Cap)
Column represents the percentage of seats filled (% capacity). Although it is incorrectly labelled as “Perfs” in the HTML source, validation against expected occupancy values confirms that it corresponds to % capacity. Since this metric directly depends on seats sold and total capacity, it is retained and renamed appropriately.




```python
df_final = df[[
    'Week Ending',
    'Avg Ticket',
    'Top Ticket',
    'Seats Sold',
    'Seats in Theatre',
    '% Cap'
]]
```
